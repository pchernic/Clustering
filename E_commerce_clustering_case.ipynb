# Step 4-5: Modeling & Model evaluation

## K-Means

K-Means is a partitioning algorithm that **divides a dataset into 'k' clusters, aiming to minimize the sum of squared distances between data points and their cluster centroids.**

```python
kelbow_visualizer(KMeans(), df_rfm_clip_scaled, k=10, timings=False)
```

```python
# Create list of dictionaries

cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results = []
X = df_rfm_clip_scaled.copy()

for k in range(2,11):
  model = KMeans(n_clusters=k, random_state=0)
  labels = model.fit_predict(X)
  cluster_results_dict = {'k': k}
  cluster_results_dict['inertia'] = model.inertia_
  for metric in cluster_metrics:
    cluster_results_dict[metric.__name__] = metric(X, labels)
  cluster_metrics_results.append(cluster_results_dict)

pd.DataFrame(cluster_metrics_results).set_index('k').style.background_gradient()
```

```python
kmeans = KMeans(4)
kmeans_labels = kmeans.fit_predict(df_rfm_clip_scaled)
```

## Gaussian Mixture (GMM)

GMM assumes that **data points are generated from a mixture of Gaussian distributions.** It estimates these distributions **to identify clusters and their associated probabilities for each data point.**


```python
# Select best model by covariance type and K (Practical Statistics for Data Scientists)
results = []
k_range = range(1,9)
covariance_types = ['full', 'tied', 'diag', 'spherical']
X = df_rfm_clip_scaled.copy()

for n_components in k_range:
  for covariance_type in covariance_types:
    mclust = GaussianMixture(n_components=n_components, n_init=10, warm_start=True, covariance_type=covariance_type)
    mclust.fit(X)
    results.append({
      'bic': mclust.bic(X),
      'n_components': n_components,
      'covariance_type': covariance_type,
    })

results = pd.DataFrame(results)
results.sort_values('bic').head()
```

```python
sns.lineplot(data=results, x='n_components', y='bic', hue='covariance_type');
```

```python
mclust = GaussianMixture(n_components=5, covariance_type='full', n_init=10)
mclust_labels = mclust.fit_predict(df_rfm_clip_scaled)
```

```python
# Adapting book  "Practical Statistics for Data Scientists" code:
fig,axes = plt.subplots(nrows=4, figsize=(14,12), sharex=True)

for i,ax in enumerate(axes):
  center = centers.loc[i,:]
  maxPC = 1.01 * center.abs().max()
  colors = ['green' if l > 0 else 'red' for l in center]
  center.plot.bar(ax=ax, color=colors)
  ax.set_ylabel(f'Cluster {i+1}')
  ax.set_ylim(-maxPC, maxPC)
  ax.axhline(color='gray')
  ax.xaxis.set_ticks_position('none')

plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

# Analysis and Insights:

 Here's a breakdown of comparisons and business insights based on the provided cluster analysis matrix:

**Comparison 1: Cluster 3 vs. Other Clusters**

- **Insight:** Cluster 3 has the lowest recency (121.06 days since the last purchase), higher frequency (3.88 purchases), and significantly higher monetary value (average spending of 80.71). These customers are the most recent, buy more frequently, and spend substantially more.

- **Business Action:** Recognize that Cluster 3 represents high-value customers. Implement personalized marketing strategies to maintain their engagement and encourage higher spending, such as exclusive offers, loyalty programs, and personalized product recommendations.

**Comparison 2: Cluster 0 vs. Cluster 1**

- **Insight:** Cluster 0 has lower recency (41.52 days) compared to Cluster 1 (67.52 days). Cluster 0 also exhibits higher frequency (10.19) and slightly lower monetary value (20.43) compared to Cluster 1 (2.70 frequency and 17.12 monetary).

- **Business Action:** Consider sending targeted reminders and promotions to customers in Cluster 0 to encourage more frequent purchases and maintain their engagement. For Cluster 1, develop strategies to increase both purchase frequency and monetary value through cross-selling or up-selling.

**Comparison 3: Cluster 2 vs. Cluster 3**

- **Insight:** Cluster 3 outperforms Cluster 2 in all aspects. Cluster 3 has a shorter recency (121.06 vs. 261.08 days), higher frequency (3.88 vs. 1.51 purchases), and significantly higher monetary value (80.71 vs. 18.10).

- **Business Action:** Understand what drives the higher spending in Cluster 3, identify the products or categories that contribute the most, and consider promoting them more prominently. For Cluster 2, work on strategies to increase both frequency and monetary value.

**Comparison 4: Cluster Sizes**

- **Insight:** Evaluate the size of each cluster to understand the customer distribution across segments.

- **Business Action:** Tailor marketing campaigns, communication, and product recommendations to each cluster's unique characteristics. Allocate resources proportionally to the size and value of each cluster for optimized engagement and revenue growth.

**Comparison 5: Time Trends**

- **Insight:** Track changes in cluster metrics over time to identify evolving customer behavior.

- **Business Action:** Use time-series analysis to adapt strategies as customer preferences shift. Implement dynamic pricing, seasonal promotions, and personalized content to align with changing trends and maximize customer engagement.

**Comparison 6: Customer Lifetime Value (CLV)**

- **Insight:** Calculate CLV for each cluster based on historical data and projected future spending.

- **Business Action:** Allocate marketing budgets and resources according to CLV. Prioritize clusters with higher CLV potential for long-term revenue growth.
